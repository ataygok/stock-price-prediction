```python
# Gerekli araçları yüklüyoruz
!pip install yfinance torch scikit-learn matplotlib

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import torch
import torch.nn as nn
import math
from sklearn.metrics import mean_squared_error

# Amazon (AMZN) verisini son 5 yıl için çekiyoruz
df = yf.download('AMZN', period='5y')
df.head()
```

```python
# 1. Kapanış fiyatlarını alıp -1 ile 1 arasına ölçeklendiriyoruz
kapanis_fiyatlari = df[['Close']].values
scaler = MinMaxScaler(feature_range=(-1, 1))
olceklendirilmis_veri = scaler.fit_transform(kapanis_fiyatlari)

# 2. Kayan pencere (sliding window) fonksiyonu
def veri_paketle(veri, pencere_boyutu):
    X, Y = [], []
    for i in range(len(veri) - pencere_boyutu - 1):
        X.append(veri[i:(i + pencere_boyutu)])
        Y.append(veri[i + pencere_boyutu])
    return np.array(X), np.array(Y)

pencere = 20
X, Y = veri_paketle(olceklendirilmis_veri, pencere)

# 3. %80 Eğitim, %20 Test olarak bölüyoruz
egitim_siniri = int(len(X) * 0.8)
X_egitim, Y_egitim = X[:egitim_siniri], Y[:egitim_siniri]
X_test, Y_test = X[egitim_siniri:], Y[egitim_siniri:]

# 4. PyTorch için Tensörlere dönüştürüyoruz
X_egitim_tensor = torch.from_numpy(X_egitim).float()
Y_egitim_tensor = torch.from_numpy(Y_egitim).float()
X_test_tensor = torch.from_numpy(X_test).float()
Y_test_tensor = torch.from_numpy(Y_test).float()

# Boyutları ayarlıyoruz [Paket Sayısı, Gün Sayısı, Özellik Sayısı]
X_egitim_tensor = torch.reshape(X_egitim_tensor, (X_egitim_tensor.shape[0], pencere, 1))
X_test_tensor = torch.reshape(X_test_tensor, (X_test_tensor.shape[0], pencere, 1))

print("Veri hazırlığı ve tensör dönüşümleri başarıyla tamamlandı!")
print(f"Eğitim Paket Sayısı: {X_egitim_tensor.shape[0]}, Test Paket Sayısı: {X_test_tensor.shape[0]}")
```

```python
# LSTM Modeli
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=32, num_layers=2, output_size=1):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).requires_grad_()
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).requires_grad_()
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))
        out = self.fc(out[:, -1, :]) 
        return out

# GRU Modeli
class GRUModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=32, num_layers=2, output_size=1):
        super(GRUModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).requires_grad_()
        out, _ = self.gru(x, h0.detach())
        out = self.fc(out[:, -1, :]) 
        return out

# Modelleri oluşturuyoruz
lstm_model = LSTMModel()
gru_model = GRUModel()

kriter = nn.MSELoss()
lstm_optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.01)
gru_optimizer = torch.optim.Adam(gru_model.parameters(), lr=0.01)

# LSTM Eğitimi
print("LSTM Eğitimi Başlıyor...")
for epoch in range(50):
    lstm_model.train()
    tahminler = lstm_model(X_egitim_tensor)
    loss = kriter(tahminler, Y_egitim_tensor)
    lstm_optimizer.zero_grad()
    loss.backward()
    lstm_optimizer.step()

# GRU Eğitimi
print("GRU Eğitimi Başlıyor...")
for epoch in range(50):
    gru_model.train()
    tahminler = gru_model(X_egitim_tensor)
    loss = kriter(tahminler, Y_egitim_tensor)
    gru_optimizer.zero_grad()
    loss.backward()
    gru_optimizer.step()

print("\nHer iki modelin antrenmanı da başarıyla tamamlandı!")
```

```python
# Değerlendirme modu
lstm_model.eval()
gru_model.eval()

with torch.no_grad():
    lstm_tahmin_tensor = lstm_model(X_test_tensor)
    gru_tahmin_tensor = gru_model(X_test_tensor)

# Orijinal fiyat birimine ($) geri dönüştürme
lstm_tahmin = scaler.inverse_transform(lstm_tahmin_tensor.numpy())
gru_tahmin = scaler.inverse_transform(gru_tahmin_tensor.numpy())
gercek_fiyatlar = scaler.inverse_transform(Y_test_tensor.numpy())

# Metriklerin hesaplanması
lstm_rmse = math.sqrt(mean_squared_error(gercek_fiyatlar, lstm_tahmin))
gru_rmse = math.sqrt(mean_squared_error(gercek_fiyatlar, gru_tahmin))

print("--- MODEL PERFORMANSLARI ---")
print(f"LSTM Modeli RMSE Skoru: {lstm_rmse:.2f}")
print(f"GRU Modeli RMSE Skoru: {gru_rmse:.2f}\n")

# Grafikleştirme
plt.figure(figsize=(12,6))
plt.plot(gercek_fiyatlar, color='black', label='Gercek Fiyat')
plt.plot(lstm_tahmin, color='red', alpha=0.7, label='LSTM Tahmini')
plt.plot(gru_tahmin, color='blue', alpha=0.7, label='GRU Tahmini')
plt.title('Amazon Hisse Senedi Tahminleri: LSTM vs GRU')
plt.xlabel('Test Gunleri')
plt.ylabel('Fiyat ($)')
plt.legend()
plt.grid(True)
plt.show()
```
